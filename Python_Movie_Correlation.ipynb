```python
# First import the packages we will use in this project

import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None
```

```python
# Read in the data and view it

df = pd.read_csv('Extracurricular/Movie Project/movies.csv')
df.head()
```

```python
# Check for null data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))   
    
# In summary we created a for loop and a variable, we tell our loop to search through all the columns within our df and print the 
# name of the column as well as the percent of null or missing values within each column in the format '{} - {}%'
```

```python
# Drop rows that have missing data

df = df.dropna()
```

```python
# Data Types for our columns

print(df.dtypes)
```

```python
# Change the data types of columns

df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
df
```

```python
# Creating a variable that identifies the correct year of release

df['yearcorrect'] = df['released'].str.extract(pat = '([0-9]{4})').astype(int)
df.head()
```

```python
# Order our Data a little bit to see

df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
# Drop any duplicating rows

df['company'].drop_duplicates().sort_values(ascending = False)
```

```python
# Build a scatter plot to test correlation hypothesis

plt.scatter(x = df['budget'], y=df['gross'])
plt.title('Budget vs Gross earnings')
plt.xlabel('Gross Earnings')
plt.ylabel('Budget for films')
plt.show()
```

```python
# Plot the budget vs gross using seaborn

sns.regplot(x='budget', y='gross', data=df, scatter_kws={'color':'red'}, line_kws={'color':'blue'})
```

```python
# Taking a look at correlation coefficients (Pearson vs Kendall vs Spearman)

df.corr(method = 'pearson')

# pearson is typically default however it is interesting to compare differneces between methods!
```

```python
df.corr(method = 'kendall')
```

```python
df.corr(method = 'spearman')
```

```python
# Visualizing correlations through a heatmap of pearson coefficents

Correlation_matrix = df.corr(method = 'pearson')
sns.heatmap(Correlation_matrix, annot = True)
plt.title('Correlation matrix of numeric columns')
plt.xlabel('Dataframe Columns')
plt.ylabel('Dataframe Columns')
plt.show()
```

```python
# create a numeric representation of the categorical variables to assess correlation

df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes

df_numerized
```

```python
# Visualizing correlations through a heatmap of pearson coefficents

Correlation_matrix = df_numerized.corr(method = 'pearson')
sns.heatmap(Correlation_matrix, annot = True)
plt.title('Correlation matrix of numeric columns')
plt.xlabel('Dataframe Columns')
plt.ylabel('Dataframe Columns')
plt.show()
```

```python
# Provide the correlation coefficient for a numerized df

df_numerized.corr()
```

```python
# Display each variable's correlation coefficients grouped by variable

correlation_matrix = df_numerized.corr()

corr_pairs = correlation_matrix.unstack()

corr_pairs
```

```python
# Create a variable to reveal correlative pairs sorted in ascending order

sorted_pairs = corr_pairs.sort_values()
sorted_pairs
```

```python
# Create a variable to reveal high correlations

high_corr = sorted_pairs[(sorted_pairs) > .5]
high_corr
```
